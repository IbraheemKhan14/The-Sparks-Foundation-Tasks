```python
# Making our correlation matrix more visualizing 
correlation_matrix = data.corr()
fig, ax = plt.subplots(figsize=(13, 8))
ax = sns.heatmap(correlation_matrix,
                 annot=True,
                 linewidths=0.5,
                 fmt=".2f",
                 cmap="YlGnBu");
plt.show()
```

```python
# Making covariance matrix
data.cov()
```

```python
# Making our covariance matrix more visualizing 
covariance_matrix = data.cov()
fig, ax = plt.subplots(figsize=(13, 8))
ax = sns.heatmap(covariance_matrix,
                 annot=True,
                 linewidths=0.5,
                 fmt=".2f",
                 cmap="YlGnBu");
plt.show()
```

## Checking the dependency for the given columns and visualizing them

```python
# Finding the the frequency of states where the items are ordered
print(data.State.value_counts())
plt.figure(figsize = (13, 8))
sns.countplot(x=data.State)
plt.xticks(rotation=90)
plt.show()
```

```python
# Finding the the frequency of different shipment modes
print(data['Ship Mode'].value_counts())
plt.figure(figsize = (13, 8))
sns.countplot(x=data['Ship Mode'])
plt.xticks(rotation=90)
plt.show()
```

```python
# Checking which categories of items are ordered the most
print(data['Sub-Category'].value_counts())
plt.figure(figsize = (13, 8))
sns.countplot(x=data['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

```python
# Checking the segment of customers ordering the most
print(data.Segment.value_counts())
plt.figure(figsize = (13, 8))
sns.countplot(x=data.Segment)
plt.xticks(rotation=90)
plt.show()
```

```python
# Sales and profits made in different regions 
plt.figure(figsize=(13,8))
sns.barplot(x=data['Region'], y=data['Sales'])
plt.xticks(rotation=90)
plt.figure(figsize=(13,8))
sns.barplot(x=data['Region'], y=data['Profit'])
plt.xticks(rotation=90)
```

```python
# Frequency of orders in different regions
print(data.Region.value_counts())
plt.figure(figsize = (13, 8))
sns.countplot(x=data.Region)
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualizing sales and profit with each of the categories
plt.figure(figsize=(13,8))
sns.barplot(x=data['Category'], y=data['Sales'])
plt.xticks(rotation=90)
plt.figure(figsize=(13,8))
sns.barplot(x=data['Category'], y=data['Profit'])
plt.xticks(rotation=90)
```

```python
# Visualizing sales and profit with each of the sub-categories
plt.figure(figsize=(13,8))
sns.barplot(x=data['Sub-Category'], y=data['Sales'])
plt.xticks(rotation=90)
plt.figure(figsize=(13,8))
sns.barplot(x=data['Sub-Category'], y=data['Profit'])
plt.xticks(rotation=90)
```

```python
# Region wise sales of categories
plt.figure(figsize=(13,8))
sns.barplot(x='Category', y='Profit', hue='Region', data=data)
plt.show()
```

```python
# Region wise distribution of shipping modes
plt.figure(figsize=(13,8))
sns.barplot(x='Ship Mode', y='Sales', hue='Region', data=data)
plt.show()
```

```python
# Plot with each of the pairs in the dataset columns
sns.pairplot(data)
```

```python
# Finding sum, mean, variance, count, standard deviation and median of profit for different states in the United States
data.groupby('State').Profit.agg(["sum","mean","min","max","count","median","std","var"])
```

## Conclusion and Observations made

* California, New York, and Texas are the states with most number of order counts
* Standard Class is the most opted class of shipping
* The segment of customers ordering the most are the "Consumers"
* Highest sales and profits are recorded in the 'Technology' category.
* Copiers generate the highest sales and profit amongst all the given sub-categories
* Even though the sales in the central region is high, but profit made is low and hence, we should find more ways to increase the profit in the central region
* The western region hosts the highest frequency of orders made.
* Even though the sales of machines are very high, but the equivalent profit is very low which is a cause of concern.
* Loss is noticed in the 'Furniture' category in the central region, relevant actions and strategies should be made to minimize the losses
